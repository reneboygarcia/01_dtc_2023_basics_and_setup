```python
#Q6 For the passengers picked up in the Astoria Zone 
# which was the drop off zone that had the largest tip? 
# We want the name of the zone, not the id.

# read csv zone
csv_zone = '/Users/reneboygarcia/Library/CloudStorage/GoogleDrive-reneboygarcia@gmail.com/My Drive/Personal/Data Science Notebook/Data Engineering/week_1_basics_and_setup/2_ingesting_ny_taxi_data_to_postgres/week1_hw/taxi+_zone_lookup.csv'
df_zone = pd.read_csv(csv_zone)
df_zone.loc[df_zone['Zone']=='Astoria']
```

```python
df_merged = pd.merge(left=df_green_tp_2019, 
                    right=df_zone,
                    how='left',
                    left_on=['DOLocationID'],
                    right_on=['LocationID']
                    )
df_merged.head()               
```

```python
cols = ['lpep_pickup_datetime','lpep_dropoff_datetime', 'PULocationID','DOLocationID','Zone', 'service_zone','tip_amount']
(df_merged[cols]
    .loc[df_merged['PULocationID']==7]
    .sort_values(by=['tip_amount'], ascending=False))

    # PULocationID == 7 is Astoria
```
